# 88x31 GIF Button Collection
## Scraper code

```python
# Import all dependencies
from bs4 import BeautifulSoup
from json import dump
from threading import Thread
import requests
import shutil
import os
import math
```

```python
# Define variables
IMAGE_BASE_URL = "https://cyber.dabamos.de/88x31/"
SITE_URLS = [
    "https://cyber.dabamos.de/88x31/index.html",
    "https://cyber.dabamos.de/88x31/index2.html",
    "https://cyber.dabamos.de/88x31/index3.html",
    "https://cyber.dabamos.de/88x31/index4.html",
]

# Variable to hold all processed buttons
buttons = []
```

```python
# Get data from the site and extract the button links
for site_url in SITE_URLS:

    # Get raw site
    site_raw = requests.get(site_url)

    # Convert to beautifulSoup
    site_bs = BeautifulSoup(site_raw.content, "html.parser")

    # Get all image elements
    site_image_elements = site_bs.findAll('img')

    # Loop through all images
    for image_element in site_image_elements:

        # Check to see if on the back button image, if so, skip it
        if image_element['alt'] == 'back':
            continue

        # Extract attributes from element, add them to a Python object
        button = {}
        button['name'] = image_element['alt']
        button['url'] = image_element['src']

        # Append the populated object to the buttons array
        buttons.append(button)
```

```python
def get_button_photos(buttons = []):
    for button in buttons:
        # Check to see if button file already exists
        if os.path.isfile('public/' + button['url']):
            print("Button ", button['name'], " already exists! Skipping...")
            continue

        # Get image from server
        img = requests.get(str(IMAGE_BASE_URL + button['url']), stream=True)

        # Save image to file
        with open('public/' + button['url'], 'wb') as f:
            shutil.copyfileobj(img.raw, f)
            print("Successfully downloaded ", button['name'])
```

```python
get_button_photos(buttons)
```

```python
# Save buttons to JSON
with open('buttons.json', 'w') as f:
    dump(buttons, f, indent=4, sort_keys=True)
```
